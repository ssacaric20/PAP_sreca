Provjerom je ustanovljeno da nema dupliciranih vrijednosti.

Sada će se pokušati popuniti nedostajuće vrijednosti korištenjem metode interpolacije, na način da će prvo retci biti kategorizirani, odnosno podijeljeni u grupe prema atributu "Country name", nakon čega bi se izračunala središnja vrijednost za danu zemlju. To će se koristiti kako bi se preciznije pogodila moguća nedostajuća vrijednost, s obzirom na odabranu državu. 

```python
hepi_podaci.reset_index(drop=True, inplace=True)

koristeci = hepi_podaci.select_dtypes(include=['float64']).columns
for column in koristeci:
 
    hepi_podaci[column] = (
    hepi_podaci.groupby('Country name')[column].apply(lambda group: group.interpolate(method='linear', limit_direction='both')).reset_index(drop=True)
        )

for column in koristeci:
    hepi_podaci[column] = hepi_podaci.groupby('Country name')[column].transform(
        lambda x: x.fillna(x.mean())
    )


print(hepi_podaci.info())
print(hepi_podaci.describe())
```

U nekim slučajevima metoda interpolacije nije bila izvediva, s obzirom da za neke države nedostaju podaci u pojedinom atributu kroz svaku godinu. Te iznimke i nedostajeće vrijednosti popunit će se fillna metodom, koja će uzimati središnju vrijednost cijelog stupca za nadopunjavanje NaN vrijednosti. 

```python
hepi_podaci = hepi_podaci.apply(lambda col: col.fillna(col.mean()) if col.dtype in ['float64'] else col)

print(hepi_podaci.info())
hepi_podaci.describe()
```

Tako su sve nedostajeće vrijednosti nadopunjene. Kao što je već spomenuto, za razliku od metode fillna, kod interpolacije su vrijednosti grupirane prema zemlji, a uzete isključivo središnje vrijednosti samo od odabrane zemlje (za razliku od metode fillna, koja koristi srednju vrijednost cijelog stupca), što je omogućilo realističnije rezultate. Ukoliko neka zemlja nije imala prikupljene podatke unutar određenog stupca, tamo su ostale NaN vrijednosti. Te su vrijednosti dodatno popunjene klasičnom fillna metodom, s obzirom da tamo nije bila primjenjiva metoda interpolacije.

Interpolacijom nisu se mijenjale krajnje vrijednosti (vrijednosti minimalnog i maksimalnog kvartila), već samo središnje, a isto vrijedi i za fillna metodu. Za razumijevanje i pomoć kod korištenja metoda interpolacije i fillna, koristio se Llego.dev [10].

Skup podataka za indeks mira:

Ovdje svaka država ima samo jedan redak, a godine (2008 do 2020) predstavljaju zasebne stupce. Svaka država prikazuje svoje vrijednosti indeksa mira za svaku godinu u posebnom stupcu. Ovaj način organiziranja podataka znači da postoji manji broj redaka jer se godina prikazuje u obliku stupaca, umjesto da postoji više redaka po godinama za svaku državu. Međutim, s obzirom na postojanje više čimbenika u prvom setu podataka, normalizirat će se skup podataka za indeks mira.

Postupak normalizacije:

Proces normalizacije podrazumijeva transformaciju trenutno svakog stupca koji predstavlja pojedinačnu godinu u zaseban redak s novim stupcima (country_name, year, peace_index). Na taj način, i indeks mira postaje organiziran u obliku koji je usporediv s indeksom sreće, gdje svaka država ima više redaka (jedan za svaku godinu).

Nakon normalizacije, oba skupa podataka (indeks sreće i indeks mira) bit će u istom formatu. Svaki redak predstavlja jednu godinu za određenu državu, a bit će jasno definiran stupac za godinu, stupac za državu i stupac za vrijednost indeksa (bilo da je to indeks sreće ili indeks mira). Na taj način bit će moguće uspoređivati ove dvije varijable kroz godine za svaku državu.

```python
con = sqlite3.connect("SQL1.db")
cursor = con.cursor()

cursor.execute("""
    CREATE TABLE IF NOT EXISTS peace_normalized (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        country_name TEXT,
        year INT,
        peace_index REAL
    )
""")
con.commit()

# potrebno je normalizirati tablicu
cursor.execute("SELECT Country, iso3c, Two8, Two9, Two10, Two11, Two12, Two13, Two14, Two15, Two16, Two17, Two18, Two19, Two20 FROM pis_file")
rows = cursor.fetchall()

normalized_data = []
years = range(2008, 2021) 

for row in rows:
    country = row[0]
    for i, year in enumerate(years):
        peace_index = row[i + 2] 
        
        
        if peace_index is not None:
            normalized_data.append((country, year, peace_index))

# normalizirani podaci
cursor.executemany("""
    INSERT INTO peace_normalized (country_name, year, peace_index)
    VALUES (?, ?, ?)
""", normalized_data)

con.commit()

#con.close()
```

## 3.2. Postavljanje Flask servera

Za razvoj API-ja, odabran je Python Flask framework. Prvo su definirane osnovne funkcionalnosti, a zatim je i postavljen sam Flask server. Na samoj početnoj stranici se korisnicima pruža informacije o dostupnim funkcijama, odnosno definiranim osnovnim rutama koje omogućuju dohvat podataka o indeksu sreće, čimbenicima sreće vezanih za pojedinačnu zemlju, filtriranje prema minimalnim i maksimalnim vrijednostima indeksa sreće, te potencijalno preklapanje između indeksa sreće i indeksa mira.

Nakon postavljanja osnovnih ruta, povezan je server s SQLite bazom podataka. To je omogućeno preko postavljenih funkcija za spajanje na bazu i izvođenje SQL upita za dohvat podataka, koji su pohranjeni u dvije tablice od kojih svaka predstavlja pojedini skup podataka. 

Za svaku funkcionalnost koja je trebala biti omogućena kroz API, definirane su odgovarajuće rute. Za početak, dohvat indeksa sreće po godinama za odabranu zemlju koji omogućava ispis popisa godina i odgovarajućih vrijednosti indeksa sreće za određenu zemlju. Zatim dohvat čimbenika koji utječu na sreću. Prilikom unosa imena zemlje, ispisali bi se svi spomenuti čimbenici i njihove vrijednosti po godini. Osim toga, moguće je i filtriranje podataka prema minimalnim i maksimalnim vrijednostima indeksa sreće. Ova funkcionalnost omogućuje korisnicima da filtriraju zemlje prema vrijednostima indeksa sreće, s mogućnosti postavljanja proizvoljnih granica. Za kraj, omogućen je pregled preklapanja između indeksa sreće i indeksa mira, fiksno postavljenim pragovima prihvatljivosti. Konkretno, "gr_sreca" postavlja minimalni indeks sreće, dok "gr_mir" određuje maksimalni indeks mira (niži indeks mira znači veći mir). Kao rezultat se vraća  popis zemalja i godina gdje je indeks sreće veći od gr_sreca, a indeks mira manji od gr_mir      . 


Nakon implementacije svih potrebnih funkcionalnosti, server je testiran kako bi se provjerilo ispravno funkcioniranje svih ruta i API poziva. Upotrebljeno je par testnih uzoraka kako bi se osiguralo da API pravilno vraća rezultate na temelju korisničkih upita. Server je pokrenut na lokalnoj adresi i portu, omogućujući korisnicima da pristupe API-ju i pretražuju podatke prema svojim potrebama. 

Za postavljanje Flask servera koristile su se bilješke s predavanja na kolegiju Programiranje za analizu podataka [7]. Za rad s dohvaćanjem argumenata preko GET metode, također su se koristile bilješke s predavanja, kao i online članak za pomoć u radu s request argumentima [7][11]. 

```python
import threading
from flask import Flask, jsonify, request
import sqlite3

app = Flask(__name__)
port = "5005"

def get_connection():
    return sqlite3.connect('SQL1.db')

# Home route
@app.route("/")
def home():
    message = (
        "Dobrodosli u Indeks Srece API! "
        "Dostupne relacije: "
        "/indeks_sreca/country?country_name=Ime "
        "/sreca/cimbenici?country_name=Ime "
        "/indeks_sreca/filter?min_index=vrijednost&max_index=vrijednost "
        "/sreca-mir/preklapanje"
    )
    return jsonify(message)

# iskljucivo indeks srece
@app.route("/indeks_sreca/country", methods=["GET"])
def get_happiness_by_country():
    country_name = request.args.get("country_name")
    if not country_name:
        return jsonify({"ERROR": "Niste dobro unjeli ime zemlje. Format: /happiness/country?country_name=Ime"}), 400

    con = get_connection()
    cursor = con.cursor()
    cursor.execute("""
        SELECT year, life_ladder 
        FROM file 
        WHERE country_name = ? 
        ORDER BY year
    """, (country_name,))
    data = cursor.fetchall()
    con.close()

    return jsonify({"Country": country_name, "Indeks srece po godini": data})

# svi cimbenici srece
@app.route("/sreca/cimbenici", methods=["GET"])
def get_happiness_factors():
    country_name = request.args.get("country_name")
    if not country_name:
        return jsonify({"ERROR": "Niste dobro unjeli ime zemlje. Format: /happiness/factors?country_name=Ime"}), 400

    con = get_connection()
    cursor = con.cursor()
    cursor.execute("""
        SELECT year, log_gdp_per_capita, social_support, healthy_life_expectancy, 
               freedom_to_make_choices, generosity, perceptions_of_corruption, 
               positive_affect, negative_affect
        FROM file 
        WHERE country_name = ? 
        ORDER BY year
    """, (country_name,))
    data = cursor.fetchall()
    con.close()

    return jsonify({"Zemlja": country_name, "Cimbenici srece": data})

# filder baziran na min i max vrijednostima indeksa srece
@app.route("/indeks_sreca/filter", methods=["GET"])
def filter_happiness_index():
    min_index = request.args.get("min_index", type=float)
    max_index = request.args.get("max_index", type=float)

    if min_index is None or max_index is None:
        return jsonify({"ERROR": "Potrebni su i min_index i max_index. Format: /happiness/filter?min_index=vrijednost&max_index=vrijednost"}), 400

    con = get_connection()
    cursor = con.cursor()
    
    cursor.execute("""
        SELECT country_name, year, life_ladder 
        FROM file 
        WHERE life_ladder BETWEEN ? AND ?
        ORDER BY life_ladder DESC
    """, (min_index, max_index))
    data = cursor.fetchall()
    con.close()

    return jsonify({"Rezultati": data})

# ovdje se uvodi drugi set podataka i potencijalno preklapanje
@app.route("/sreca-mir/preklapanje", methods=["GET"])
def preklapanje():
    gr_sreca = request.args.get("gr_sreca", type=float, default=5.0)
    gr_mir = request.args.get("gr_mir", type=float, default=1.5)
    
    con = get_connection()
    cursor = con.cursor()
    
    cursor.execute("""
        SELECT h.country_name, h.year, h.life_ladder, p.peace_index
        FROM file h
        INNER JOIN peace_normalized p
        ON h.country_name = p.country_name AND h.year = p.year
        WHERE h.life_ladder > ? AND p.peace_index < ?
        ORDER BY h.year, h.country_name
    """, (gr_sreca, gr_mir))
    
    data = cursor.fetchall()
    con.close()

    return jsonify({"Uocena preklapanja": data})


# start flask servera
threading.Thread(target=app.run, kwargs={"host": "127.0.0.1", "port": port, "use_reloader": False}).start()

#app.run(host="127.0.0.1", port=port, debug=True)
```

## 3.3. Vizualizacija podataka

U sljedećem dijelu projekta nastojit će se obraditi vizualizacija podata, a provest će se putem biblioteka za grafičko prikazivanje podataka Seaborn, odnosno Pyplot. Konkretnije, vizualizacija će se fokusirati na tri osnovna grafa; boxplot, histogram, i heatmap, od kojih svaki ima specifičnu ulogu i svrhu kod analite podataka. 

Za početak, objasnit će se pojam boxplot dijagrama. Boxplot dijagram je grafički prikaz raspodjele podataka, gdje je glavna poanta prikazati medijane i kvartile, a također omogućuje i lako prepoznavanje stršila. Dolje obojane kutije označavaju interkvartilni raspon, odnosno podatke između prvog i trećeg kvartila, dok gornji i donji "brkovi" označavaju totalni raspon podataka. Stršila, dosnosno outliers su predstavljeni kao točke van spomenutih brkova. Sam medijan se unutar kutije označava horizontalnom linijom, a predstavlja najučestaliju vrijednost.

U ovom slučaju, prikazivat će se distribucija indeksa sreće kroz godine. S obzirom na to kako su dijagrami poredani jedan za drugim, moguće je lako pratiti promjene kroz vrijeme, kao i moguća stršila.

Za izvedbu svih navedenih dijagrame koristili su se materijali iz laboratorijskih vježbi sa kolegija Uvod u umjetnu inteligenciju [9]. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.boxplot(data=hepi_podaci, x='year', y='Life Ladder')
plt.title('Indeks srece kroz godine')
plt.xlabel('Godine')
plt.ylabel('Indeks srece')
plt.show()
```

Vidjlivo je da osim iznimke u prijelazu između 2005. i 2006. godine, indeks sreće ukazuje na jako blagi pozitivan rast, dok su stršila primjetiva tek početkom 2019. godine. Iz ovog prikaza je moguće zaključiti da makar je promjena jako blaga, dolazi do pozitivnog rasta. Osim toga, vrijedi napomenuti da kod većine ispitanih zemalja indeks sreće prelazi vrijednost od 5, što bi impliciralo da se većina stanovništva osjeća relativno zadovoljno. 

Kao što je već napomenuto, do 2019. godine nema stršila, što ukazuje na relativnu stabilnost među državama u kontekstu sreće i ekonomskih indikatora. Međutim, s obzirom na to da je pandemija COVID-19 imala globalne posljedice, osobito na zdravlje, ekonomiju i društvene sfere života, moglo bi se pretpostaviti da je iz tog razloga došlo do značajnih padova u ekonomskom stanju nekih država, i da je to razlog zašto je točno te godine došlo do pojave ekstremnih vrijednosti u podacima. Moguće je da su u ekonomskim čimbenicima poput socijalne podrške i BDP-a došle nagle promjene, što bi automatski utjecalo na ispitani indeks sreće.

Nadalje, prikazat će se potencijalne korelacije među podacima koristeći heatmap; vizualni prikaz matrice podataka, gdje se intenzitet boje koristi za označavanje vrijednosti.

Heatmap će se koristiti za analizu povezanosti između prije navedeni čimbenika: indeks sreće, BDP, društvena podršla, velikodušnost, itd. Cilj je identificirati snažne pozitivne ili negativne korelacije između varijabli koje bi mogle objasniti obrasce u podacima.

Kako bi se iščitali podaci, napomenut će se da tamnije boje predstavljaju jače korelacije (pozitivne ili negativne), tj. korelacija bliska 1 ili -1 ukazuje na snažnu vezu, dok korelacija bliska 0 sugerira slab odnos. Tople nijanse prikazuju pozitivan utjecaj, dok hladnije nijanse ukazuju na negativan učinak.


```python
hp2 = hepi_podaci.drop(columns=['Country name', 'year'])

plt.figure(figsize=(12, 8))
correlation_matrix = hp2.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap of Happiness Factors')
plt.show()
```

Iz gore vidljivog prikaza, jasno je vidljiva razlika u čimbenicima koji pozitivno, odnosno negativno međusobno utječu jedni na druge. 

S obzirom da je glavna ideja analizirati čimbenike koji utječu na indeks sreće specifično, koristit će se prvi redak za referencu. Uočljivo je da najznačajniji pozitivan utjecaj na indeks sreće imaju BDP, socijalna podrška, i zdravlje, odnosno očekivano trajanje života, koje ovisi o zdravstvenom sustavu i skrb koju država pruža pojedincu. Vrijednost veća od 0.7 ukazuje na relativno značajnu korelaciju, što znači da bi povećanje i jednog od tog kriterija automatski i povećalo sreću pojedinca. Nešto manji utjecaj imaju čimbenici slobode donošenja odluka i pozitivnog učinka, na početku objašnjen kao "iskušenje pozitivnih emocija ispitanika u prethodnom danu", u vrijednosti oko 0.5. 

Za sad su navedeni čimbenici koji su pozitivno utjecali na indeks sreće, a sada slijede oni koji negativno utječu na njega. Makar ne toliko značajni kao i prva tri prethodno spomenuta, negativan utjecaj ima percepcija korupcije, mišljenja ispitanika o postojanju korupcije unutar vlade i poslovanja, te negativan učinak, prethodno objašnjen kao "iskušenje negativnih emocija ispitanika u prethodnom danu". Jasno je zašto bi korupcija unutar države negativno utjecala na doživljaj sreće njezinih stanovnika; korumpirani sustavi često onemogućuju zadovoljenje osnovnih potreba, čineći svakodnevni život težim.
